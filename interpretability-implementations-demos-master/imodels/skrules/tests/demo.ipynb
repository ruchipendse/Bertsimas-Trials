```python
from pyoptree.optree import OptimalTreeModel
feature_names = np.array(["x1", "x2"])
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
#X = np.array([[1, 20, 20, 20, 33, 99], [19, 27, 15, 60, 31, 20]]).T
#y = np.array([1, 1, 0, 0, 0, 1]).reshape(-1, 1)
#X_test = np.array([[1, 1, 2, 2, 2, 3, 3, 4], [1, 2, 2, 1, 0, 1, 0, 1]]).T
#y_test = np.array([1, 1, 1, 0, 0, 0, 0, 1])

#X = np.random.randn(10, 10)
X, y = make_classification(n_samples = 1000, n_features=7, n_redundant=0, n_informative=2, random_state = 0, weights=[0.7, 0.3])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test

model = OptimalTreeModel(tree_depth=3, N_min=1, alpha=0.1) #, solver_name='baron'
model.fit(X_train, y_train) # this method is currently using the fast, but not optimal solver
preds = model.predict(X_test)
#print(np.mean())
print(preds)
print('acc', np.mean(preds == y_test))
```
